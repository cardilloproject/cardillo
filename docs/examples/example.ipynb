Jupyter notebook example
========================

Let's print some numpy array

````python
import numpy as np

print(np.random.rand(2, 3, 2))
````

or visualize some data

````python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), "-")
plt.plot(x, np.cos(x), "--")
````

```{note}
In order to get this working with PyNSD we have to supply the package to the pip package manager. Otherwise we cannot load the package while building the documentation. However, it might be possible to make a local installation before building the docs.
```
